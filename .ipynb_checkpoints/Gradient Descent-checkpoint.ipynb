## Slope and derivatives

Challenge : Create a python function for the derivative of $f(x)$ called $df(x)$

```python
def df(x) : 
    return 2*x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('f(x)',fontsize = 16)
plt.plot(x_1,f(x_1), color = "blue", linewidth = 3)

# Second chart : Derivative
plt.subplot(1, 2, 2)

plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)
```

```python
# Plot function and derivative top to bottom
plt.figure(figsize=[10, 15])

# First chart : Cost function
plt.subplot(2, 1, 1)    

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('f(x)',fontsize = 16)
plt.plot(x_1,f(x_1), color = "blue", linewidth = 3)

# Second chart : Derivative
plt.subplot(2, 1, 2)

plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)
```

# Python loops and gradient descent

```python
# Gradient descent
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.0001
range_of_loop = 41

#lists for storing the values of x and slopes
x_list = [new_x]
slope_list = [df(new_x)]

for n in range(range_of_loop) : 
    prev_x = new_x 
    gradient = df(prev_x)
    new_x = prev_x - step_multiplier * gradient
    #print('New x at iteration ',n,' is: ',new_x)
    #if the new_x and the prev_x are getting closer and closer together, then the calculation (differce) 
    #between them is getting more precise 
    
    step_size = abs(new_x - prev_x)
    #print(step_size)
    
    #if the step size is less than precision, stop the loop because the desired precision has been achieved
    if step_size < precision : 
        print('The loop broke at iteration no. ', n)
        break
    elif n == range_of_loop - 1:
        print('The iteration never has never stopped because the number of times it ran was too small for the precision of the result to be achieved. ')
    
    #append desired values to their specific list
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
print('Local minimum occurs at: ', new_x)
print('Slope or df(x) value at this point is: ', df(new_x))
print('f(x) value or cost at this point is: ', f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot 

plt.figure(figsize=[20, 5])

    # First chart : Cost function
plt.subplot(1, 3, 1)    

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('f(x)',fontsize = 16)
plt.plot(x_1,f(x_1), color = "blue", linewidth = 3)

    # Scatter chart with values of x

#type conversion because the function can not work with lists
array_of_x_list_values = np.array(x_list)

plt.scatter(x_list, f(array_of_x_list_values), color='red', s=100, alpha=0.6)

# Second chart : Derivative (slope)
plt.subplot(1, 3, 2)

plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)

# Scatter chart with values of slope

plt.scatter(x_list, slope_list, color='black', s=100, alpha=0.6)

    # Third chart : Derivative (close up)
    
plt.subplot(1, 3, 3)

plt.grid()
plt.title('Gradient descent (close up)')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.xlim(-0.55,-0,2)
plt.ylim(-0.3,0.8)
plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)

    # Scatter chart with values of slope

plt.scatter(x_list, slope_list, color='black', s=100, alpha=0.6)
```

# Example 2  - Multiple minimal vs Initial guess & Advanced functions


## $$g(x) = x^4 - 4x^2 + 5$$


```python
# Making some data

x_2 = np.linspace(-2,2,1000)

# Challenge : Write the g(x) and dg(x) functions in Python.

def g(x) : 
    return x**4 - 4*x**2 + 5

def dg(x) :
    return 4*x**3 - 8*x 
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim(-2,2)
plt.ylim(0.5,5.5)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('g(x)',fontsize = 16)
plt.plot(x_2,g(x_2), color = "blue", linewidth = 3)

# Second chart : Derivative
plt.subplot(1, 2, 2)

plt.xlim(-2,2)
plt.ylim(-6,8)
plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('dg(x)',fontsize = 16)
plt.plot(x_2,dg(x_2), color  = 'orange', linewidth = 5)
```

## Gradient descent as a python function

```python
# first 2 arguments are required, next 2 are optional because they have default values
def gradient_descent(derivative_func, initial_guess, range_of_loop, multiplier=0.02, precision=0.001) :
        
    new_x = initial_guess
    
    #lists for storing the values of x and slopes
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(range_of_loop) : 
        try:
            prev_x = new_x 
            gradient = derivative_func(prev_x)
            new_x = prev_x - multiplier * gradient
            #print('New x at iteration ',n,' is: ',new_x)
            #if the new_x and the prev_x are getting closer and closer together, then the calculation (differce) 
            #between them is getting more precise 

            step_size = abs(new_x - prev_x)

            #append desired values to their specific list
            slope_list.append(derivative_func(new_x))
            x_list.append(new_x)

            #if the step size is less than precision, stop the loop because the desired precision has been achieved
            if step_size < precision : 
                print('The loop broke at iteration no. ', n, ' which means that the list of derivatives has ', n + 2, ' elements because of the value that has been given when this list was initialised and because the iteration starts at 0')
                break
            elif n == range_of_loop - 1:
                print('The iteration has never stopped because the number of times it ran was too small for the precision of the result to be achieved. ')
        except OverflowError : 
            print('An Overflow Error has occured. That means the result is too large. ')
            print(len(x_list),len(slope_list))
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, -0.5, 100)

print('Local min occurs at: ', local_min)
print('Number of steps : ', len(list_x))
```

# Example 3 - Divergence, overflow 

## $$h(x) = x^5 - 2x^4 + 2$$


```python
#Generate data 

x_3 = np.linspace(start=-2.5,stop=2.5, num=1000)

def h(x) : 
    return x**5 - 2*x**4 + 2
def dh(x) : 
    return 5*x**4 - 8*x**3 
```

```python
#Calling gradient descent function 

local_min, list_x, deriv_list = gradient_descent(derivative_func = dh,initial_guess = -0.2, range_of_loop=70)

# Plot function, derivative and scatter plot side by side
plt.figure(figsize=[15, 5])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim(-1.2,2.5)
plt.ylim(-1,4)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('h(x)',fontsize = 16)

plt.plot(x_3,h(x_3), color = "blue", linewidth = 3,alpha=0.8)

# Scatter plot 1 of gradient descent values
plt.scatter(list_x,h(np.array(list_x)), color='red',s=100,alpha=0.6)

# Second chart : Derivative 
plt.subplot(1, 2, 2)

plt.xlim(-1,2)
plt.ylim(-4,5)
plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('dh(x)',fontsize = 16)
plt.plot(x_3,dh(x_3), color  = 'orange', linewidth = 5,alpha=0.8)

# Scatter plot 2 of gradient descent values

plt.scatter(list_x, deriv_list, color='black',s=100,alpha=0.6)

print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
len(deriv_list)
```

# Learning rate

```python
#Calling gradient descent function 

local_min, list_x, deriv_list = gradient_descent(dg, 1.9, 1000, multiplier=0.25)

# Plot function, derivative and scatter plot side by side
plt.figure(figsize=[15, 5])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim(-2,2)
plt.ylim(0.5,5.5)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('g(x)',fontsize = 16)

plt.plot(x_2,g(x_2), color = "blue", linewidth = 3,alpha=0.8)

# Scatter plot 1 of gradient descent values
plt.scatter(list_x,g(np.array(list_x)), color='red',s=100,alpha=0.6)

# Second chart : Derivative
plt.subplot(1, 2, 2)

plt.xlim(-2,2)
plt.ylim(-6,8)
plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('dg(x)',fontsize = 16)
plt.plot(x_2,dg(x_2), color  = 'orange', linewidth = 5,alpha=0.8)

# Scatter plot 2 of gradient descent values

plt.scatter(list_x, deriv_list, color='black',s=100,alpha=0.6)

print("Number of steps: ", len(list_x))
```

Note : For a big multiplier, the algorithm isn't converging; it is just going and going. It also can overflow, if the learning rate is to big (the multiplier). 

```python
#Run gradient descent 3 times
iterations = 100

low_gamma = gradient_descent(derivative_func=dg, initial_guess=3.0, range_of_loop=iterations, multiplier=0.0005, precision=0.0001)

mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3.0, range_of_loop=iterations, multiplier=0.001, precision=0.0001)

high_gamma = gradient_descent(derivative_func=dg, initial_guess=3.0, range_of_loop=iterations, multiplier=0.0025, precision=0.0001)

crazy_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, range_of_loop=iterations, multiplier=0.25, precision=0.0001)

# Plot function, derivative and scatter plot side by side
plt.figure(figsize=[20, 10])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim(0,iterations)
plt.ylim(0,50)

plt.title('Effect of the learning rate ', fontsize=17)
plt.xlabel('Nr. of iterations',fontsize = 16)
plt.ylabel('Cost',fontsize = 16)

# Values for our charts
# 1) Y axis data: convert the list to numpy array

low_values = np.array(low_gamma[1])
mid_values = np.array(mid_gamma[1])
high_values = np.array(high_gamma[1])
crazy_values = np.array(crazy_gamma[1])

# 2) x axis data: create a list from 0 to iterations+1 (extra guess means "+1")

iterations_list = list(range(0,iterations+1))


print(len(low_values), len(iterations_list))

# Plotting low learning rate
plt.plot(iterations_list, g(low_values),color='lightgreen',linewidth=5, zorder=1)
plt.scatter(iterations_list, g(low_values),color='black',linewidth=1, alpha=0.8, zorder=2)

# Plotting middle learning rate
plt.plot(iterations_list, g(mid_values),color='pink',linewidth=5, zorder=1)
plt.scatter(iterations_list, g(mid_values),color='blue',linewidth=1, alpha=0.8, zorder=2)

# Plotting high learning rate

plt.plot(iterations_list, g(high_values),color='grey',linewidth=5, zorder=1)
plt.scatter(iterations_list, g(high_values),color='yellow',linewidth=1, alpha=0.8, zorder=2)

# Plotting crazy learning rate

plt.plot(iterations_list, g(crazy_values),color='red',linewidth=5, zorder=1)
plt.scatter(iterations_list, g(crazy_values),color='red',linewidth=1, alpha=0.8, zorder=2)

plt.show()
```

## Conclusion : the highest learning rate converges the fastest.

Note : Higher multipliers don't always work out, as you can see.
