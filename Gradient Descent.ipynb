## Slope and derivatives

Challenge : Create a python function for the derivative of $f(x)$ called $df(x)$

```python
def df(x) : 
    return 2*x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('f(x)',fontsize = 16)
plt.plot(x_1,f(x_1), color = "blue", linewidth = 3)

# Second chart : Derivative
plt.subplot(1, 2, 2)

plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)
```

# Python loops and gradient descent

```python
# Gradient descent
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.0001
range_of_loop = 41

#lists for storing the values of x and slopes
x_list = [new_x]
slope_list = [df(new_x)]

for n in range(range_of_loop) : 
    prev_x = new_x 
    gradient = df(prev_x)
    new_x = prev_x - step_multiplier * gradient
    #print('New x at iteration ',n,' is: ',new_x)
    #if the new_x and the prev_x are getting closer and closer together, then the calculation (differce) 
    #between them is getting more precise 
    
    step_size = abs(new_x - prev_x)
    #print(step_size)
    
    #if the step size is less than precision, stop the loop because the desired precision has been achieved
    if step_size < precision : 
        print('The loop broke at iteration no. ', n)
        break
    elif n == range_of_loop - 1:
        print('The iteration never has never stopped because the number of times it ran was too small for the precision of the result to be achieved. ')
    
    #append desired values to their specific list
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
print('Local minimum occurs at: ', new_x)
print('Slope or df(x) value at this point is: ', df(new_x))
print('f(x) value or cost at this point is: ', f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot 

plt.figure(figsize=[20, 5])

    # First chart : Cost function
plt.subplot(1, 3, 1)    

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('f(x)',fontsize = 16)
plt.plot(x_1,f(x_1), color = "blue", linewidth = 3)

    # Scatter chart with values of x

#type conversion because the function can not work with lists
array_of_x_list_values = np.array(x_list)

plt.scatter(x_list, f(array_of_x_list_values), color='red', s=100, alpha=0.6)

# Second chart : Derivative (slope)
plt.subplot(1, 3, 2)

plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)

# Scatter chart with values of slope

plt.scatter(x_list, slope_list, color='black', s=100, alpha=0.6)

    # Third chart : Derivative (close up)
    
plt.subplot(1, 3, 3)

plt.grid()
plt.title('Gradient descent (close up)')
plt.xlabel('x',fontsize = 16)
plt.ylabel('df(x)',fontsize = 16)
plt.xlim(-0.55,-0,2)
plt.ylim(-0.3,0.8)
plt.plot(x_1,df(x_1), color  = 'orange', linewidth = 5)

    # Scatter chart with values of slope

plt.scatter(x_list, slope_list, color='black', s=100, alpha=0.6)
```

## $$g(x) = x^4 - 4x^2 + 5$$


```python
# Making some data

x_2 = np.linspace(-2,2,1000)

# Challenge : Write the g(x) and dg(x) functions in Python.

def g(x) : 
    return x**4 - 4*x**2 + 5

def dg(x) :
    return 4*x**3 - 8*x 
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# First chart : Cost function
plt.subplot(1, 2, 1)    

plt.xlim(-2,2)
plt.ylim(0.5,5.5)

plt.title('Cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('g(x)',fontsize = 16)
plt.plot(x_2,g(x_2), color = "blue", linewidth = 3)

# Second chart : Derivative
plt.subplot(1, 2, 2)

plt.xlim(-2,2)
plt.ylim(-6,8)
plt.grid()
plt.title('Slope of the cost function ')
plt.xlabel('x',fontsize = 16)
plt.ylabel('dg(x)',fontsize = 16)
plt.plot(x_2,dg(x_2), color  = 'orange', linewidth = 5)
```

## Gradient descent as a python function

```python
# first 2 arguments are required, next 2 are optional because they have default values
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001) :
    
    
    new_x = initial_guess
    range_of_loop = 41

    #lists for storing the values of x and slopes
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(range_of_loop) : 
        prev_x = new_x 
        gradient = derivative_func(prev_x)
        new_x = prev_x - multiplier * gradient
        #print('New x at iteration ',n,' is: ',new_x)
        #if the new_x and the prev_x are getting closer and closer together, then the calculation (differce) 
        #between them is getting more precise 

        step_size = abs(new_x - prev_x)
        
        #append desired values to their specific list
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        

        #if the step size is less than precision, stop the loop because the desired precision has been achieved
        if step_size < precision : 
            print('The loop broke at iteration no. ', n)
            break
        elif n == range_of_loop - 1:
            print('The iteration never has never stopped because the number of times it ran was too small for the precision of the result to be achieved. ')

    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, -0.5, 0.02, 0.001)

print('Local min occurs at: ', local_min)
print('Number of steps : ', len(list_x))
```
